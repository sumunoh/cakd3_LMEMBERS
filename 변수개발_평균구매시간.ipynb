```python
time_1_count = time_1.groupby(['고객번호'])['제휴사'].agg(**{
    '1_구매갯수':lambda x:round(x.count())
}).reset_index()

time_2_count = time_2.groupby(['고객번호'])['제휴사'].agg(**{
    '2_구매갯수':lambda x:round(x.count())
}).reset_index()

time_3_count = time_3.groupby(['고객번호'])['제휴사'].agg(**{
    '3_구매갯수':lambda x:round(x.count())
}).reset_index()

time_4_count = time_4.groupby(['고객번호'])['제휴사'].agg(**{
    '4_구매갯수':lambda x:round(x.count())
}).reset_index()
```
