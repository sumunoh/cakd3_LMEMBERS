```python
s = ['1P_가구/가전/전자기기_rank', '1P_기타_rank',
       '1P_반려/취미/문구/도서_rank', '1P_생활/건강_rank', '1P_스포츠/레저/자동차_rank',
       '1P_식품_rank', '1P_의류/잡화_rank', '1P_출산유아동_rank']      
```

```python
def w(x,i):
  s9 = results[s[i]].max()
  s2 = s9/9
  if x==0:
    re = 1
  elif x < s2:
    re = 10
  elif x < 2*s2:
    re = 9
  elif x < 3*s2:
    re = 8
  elif x < 4*s2:
    re = 7
  elif x < 5*s2:
    re = 6
  elif x < 6*s2:
    re = 5
  elif x < 7*s2:
    re = 4
  elif x < 8*s2:
    re = 3
  else:
    re = 2
  return re
```

```python
results['1P_가구/가전/전자기기_r']=results[s[0]].apply(lambda x: w(x,0))
results['1P_기타_r']=results[s[1]].apply(lambda x: w(x,1))
results['1P_반려/취미/문구/도서_r']=results[s[2]].apply(lambda x: w(x,2))
results['1P_생활/건강_r']=results[s[3]].apply(lambda x: w(x,3))
results['1P_스포츠/레저/자동차_r']=results[s[4]].apply(lambda x: w(x,4))
results['1P_식품_r']=results[s[5]].apply(lambda x: w(x,5))
results['1P_의류/잡화_r']=results[s[6]].apply(lambda x: w(x,6))
results['1P_출산유아동_r']=results[s[7]].apply(lambda x: w(x,7))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif(data):
  vif = []
  data_val = data.values
  for i in range(len(data.columns)):
    vif.append(variance_inflation_factor(data_val, i))
  vif_dataframe = pd.DataFrame()
  vif_dataframe['columns'] = data.columns
  vif_dataframe['VIF'] = vif
  return vif_dataframe
```
