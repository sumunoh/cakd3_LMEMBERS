```python
again = pd.read_csv('/content/drive/MyDrive/LMEMBERS 프로젝트(1차)/dataset/1015/again_dataset.csv',encoding='cp949',index_col=0)
```

```python
a1 = again['구매금액_A_1'].sum()
a2 = again['구매금액_A_2'].sum()
a3 = again['구매금액_A_3'].sum()
a4 =again['구매금액_A_4'].sum()

b1 = again['구매금액_B_1'].sum()
b2 = again['구매금액_B_2'].sum()
b3 = again['구매금액_B_3'].sum()
b4 = again['구매금액_B_4'].sum()

c1 = again['구매금액_C_1'].sum()
c2 = again['구매금액_C_2'].sum()
c3 = again['구매금액_C_3'].sum()
c4 = again['구매금액_C_4'].sum()

d1 = again['구매금액_D_1'].sum()
d2 = again['구매금액_D_2'].sum()
d3 = again['구매금액_D_3'].sum()
d4 = again['구매금액_D_4'].sum()
```

```python
제휴사별기수별매출 = pd.DataFrame({'quarter':['14-1','14-2','15-1','15-2'],
                   'A':[a1, a2, a3, a4],
                   'B':[b1, b2, b3, b4],
                   'C':[c1, c2, c3, c4],
                   'D':[d1, d2, d3, d4]
                   })
```

```python
제휴사별기수별매출.set_index('quarter',inplace=True)
```

```python
제휴사별기수별매출
```

```python
제휴사별기수별매출.plot(kind='bar',figsize=(10,8))
plt.title( '제휴사별 매출변화', size=20 )
plt.ylabel('매출 합계')
plt.xlabel('기수')
plt.legend(labels=제휴사별기수별매출.columns, loc='best')
plt.show()
```

```python
cust_demo = pd.read_csv('/content/drive/MyDrive/LMEMBERS 프로젝트(1차)/dataset/cust_demo.csv')
```

```python
cust_demo.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
cust_demo.head()
```

```python
data2 = pd.merge(data,cust_demo,on='고객번호',how='outer')
```

```python
data2.head()
```

```python
data2 = data2[['고객번호','대분류코드','중분류코드','구매금액']]
```

```python
data2 = data2.astype({'대분류코드':'str','중분류코드':'str'})
data2.dtypes
```

```python
aggregations={'구매금액':'sum','대분류코드':lambda x:x.mode(),'중분류코드':lambda x:x.mode()}

data_eda = data2.groupby('고객번호').agg(aggregations).reset_index()
data_eda.to_csv('/content/drive/MyDrive/LMEMBERS 프로젝트(1차)/dataset_bmcat_mode',encoding='cp949')
```

```python
data4 = pd.merge(data_eda,cust_demo,on='고객번호',how='outer')
data4.head()
```

```python
data4 = data4[['고객번호','성별','연령대','대분류코드','중분류코드','구매금액']]
data4.head()
```

```python
data5 = data4.rename(columns={'고객번호':'custid','성별':'sex','연령대':'age','대분류코드':'cat_b_mode','중분류코드':'cat_m_mode','구매금액':
                              'amount'})
data5.head()
```

```python
data5 = data5.astype({'cat_b_mode':'str','cat_m_mode':'str'})
```

```python
# 남성이 많이 사는 카테고리 순위(대분류,중분류)
man_procat_mode = data5[data5['sex']=='M']['cat_b_mode'].value_counts()
```

```python
man_procat_mode
```

```python
man_procat_mode.plot(kind='bar', figsize=(10,8))
```

```python
data5.to_csv('/content/drive/MyDrive/LMEMBERS 프로젝트(1차)/고객별_대분류중분류(최빈).csv',encoding='cp949')
```

```python
p1 = pd.pivot_table(data5,values='amount',index='sex',columns='age',aggfunc='sum')
```

```python
p1.plot(kind='bar', figsize=(10,8))
plt.title( '성별 및 연령대별 매출 합계', size=20 )
plt.ylabel('매출 합계')
plt.xlabel('성별')
plt.legend(labels=p1.columns, loc='best')
plt.show()
```
