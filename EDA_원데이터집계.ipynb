```python
# 고객별 증감율

data_low2['1p'] = (data_low2['14년 2분기'] - data_low2['14년 1분기']) / data_low2['14년 1분기']
data_low2['2p'] = (data_low2['14년 3분기'] - data_low2['14년 2분기']) / data_low2['14년 2분기']
data_low2['3p'] = (data_low2['14년 4분기'] - data_low2['14년 3분기']) / data_low2['14년 3분기']
data_low2['4p'] = (data_low2['15년 1분기'] - data_low2['14년 4분기']) / data_low2['14년 4분기']
data_low2['5p'] = (data_low2['15년 2분기'] - data_low2['15년 1분기']) / data_low2['15년 1분기']
data_low2['6p'] = (data_low2['15년 3분기'] - data_low2['15년 2분기']) / data_low2['15년 2분기']
data_low2['7p'] = (data_low2['15년4분기'] - data_low2['15년 3분기']) / data_low2['15년 3분기']
```

```python
# 전체적인 증감율

dea1 = (data_low2['14년 2분기'].mean() - data_low2['14년 1분기'].mean()) / data_low2['14년 1분기'].mean()
dea2 = (data_low2['14년 3분기'].mean() - data_low2['14년 2분기'].mean()) / data_low2['14년 2분기'].mean()
dea3 = (data_low2['14년 4분기'].mean() - data_low2['14년 3분기'].mean()) / data_low2['14년 3분기'].mean()
dea4 = (data_low2['15년 1분기'].mean() - data_low2['14년 4분기'].mean()) / data_low2['14년 4분기'].mean()
dea5 = (data_low2['15년 2분기'].mean() - data_low2['15년 1분기'].mean()) / data_low2['15년 1분기'].mean()
dea6 = (data_low2['15년 3분기'].mean() - data_low2['15년 2분기'].mean()) / data_low2['15년 2분기'].mean()
dea7 = (data_low2['15년4분기'].mean() - data_low2['15년 3분기'].mean()) / data_low2['15년 3분기'].mean()


de1 = (data_low2['14년 2분기'].median() - data_low2['14년 1분기'].median()) / data_low2['14년 1분기'].median()
de2 = (data_low2['14년 3분기'].median() - data_low2['14년 2분기'].median()) / data_low2['14년 2분기'].median()
de3 = (data_low2['14년 4분기'].median() - data_low2['14년 3분기'].median()) / data_low2['14년 3분기'].median()
de4 = (data_low2['15년 1분기'].median() - data_low2['14년 4분기'].median()) / data_low2['14년 4분기'].median()
de5 = (data_low2['15년 2분기'].median() - data_low2['15년 1분기'].median()) / data_low2['15년 1분기'].median()
de6 = (data_low2['15년 3분기'].median() - data_low2['15년 2분기'].median()) / data_low2['15년 2분기'].median()
de7 = (data_low2['15년4분기'].median() - data_low2['15년 3분기'].median()) / data_low2['15년 3분기'].median()
```

```python
data_low2['1p_after'] = data_low2['1p'] / de1
data_low2['2p_after'] = data_low2['2p'] / de2
data_low2['3p_after'] = data_low2['3p'] / de3
data_low2['4p_after'] = data_low2['4p'] / de4
data_low2['5p_after'] = data_low2['5p'] / de5
data_low2['6p_after'] = data_low2['6p'] / de6
data_low2['7p_after'] = data_low2['7p'] / de7

data_low2['1p_before'] = data_low2['1p'] / dea1
data_low2['2p_before'] = data_low2['2p'] / dea2
data_low2['3p_before'] = data_low2['3p'] / dea3
data_low2['4p_before'] = data_low2['4p'] / dea4
data_low2['5p_before'] = data_low2['5p'] / dea5
data_low2['6p_before'] = data_low2['6p'] / dea6
data_low2['7p_before'] = data_low2['7p'] / dea7
```

```python
data_low2['14_1'] = data_low2['14년 1분기'] + data_low2['14년 2분기']
data_low2['14_2'] = data_low2['14년 3분기'] + data_low2['14년 4분기']
data_low2['15_1'] = data_low2['15년 1분기'] + data_low2['15년 2분기']
data_low2['15_2'] = data_low2['15년 3분기'] + data_low2['15년4분기']
```
