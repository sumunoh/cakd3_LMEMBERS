```python
def virtualize(x,y,data):
  plt.figure(figsize=(20,10))
  sns.barplot(x=x,y=y,data=data,palette='Accent')

def groupping_sum(x,y,data): # groupby_sum 적용 / groupping('연령대','구매금액',연령대와 구매금액을 포함한 데이터 프레임)
  result = data.groupby(x)[[y]].sum()
  result.reset_index(inplace=True)
  result.sort_values(by=y,ascending=False,inplace=True)
  return result

def groupping_mean(x,y,data): # groupby_mean 적용 / groupping('연령대','구매금액',연령대와 구매금액을 포함한 데이터 프레임)
  result = data.groupby(x)[[y]].mean()
  result.reset_index(inplace=True)
  result.sort_values(by=y,ascending=False,inplace=True)
  return result

def groupping_count(x,y,data): # groupby_count 적용 / groupping('연령대','구매금액',연령대와 구매금액을 포함한 데이터 프레임)
  result = data.groupby(x)[[y]].count()
  result.reset_index(inplace=True)
  result.sort_values(by=y,ascending=False,inplace=True)
  return result

def groupping_max(x,y,data): # groupby_max 적용 / groupping('연령대','구매금액',연령대와 구매금액을 포함한 데이터 프레임)
  result = data.groupby(x)[[y]].max()
  result.reset_index(inplace=True)
  result.sort_values(by=y,ascending=False,inplace=True)
  return result

def auto_virtualize(x,y,data,aggument):
  if aggument == 'sum':
    result = groupping_sum(x,y,data)
    virtualize(x,y,result)

  elif aggument =='mean':
    result = groupping_mean(x,y,data)
    print('mean')
    virtualize(x,y,result)

  elif aggument =='count':
    result = groupping_count(x,y,data)
    print('count')
    virtualize(x,y,result)

  elif aggument =='max':
    result = groupping_max(x,y,data)
    print('max')
    virtualize(x,y,result)

  else :
    print('잘못된 aggument')
```
